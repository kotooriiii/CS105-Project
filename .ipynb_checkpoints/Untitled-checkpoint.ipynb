```python
import pandas as pd

#Read basics.tsv
basics_df = pd.read_csv("basics.tsv", sep="\t")

#Filter
basics_df = basics_df[~basics_df["titleType"].isin(["tvEpisode","short", "tvSeries", "tvMiniSeries", "video"])]

basics_df = basics_df[basics_df["genres"] != "\\N"]
basics_df = basics_df[basics_df["startYear"] != "\\N"]
basics_df["startYear"] = basics_df["startYear"].apply(int)
basics_df = basics_df[basics_df["startYear"] > 1960]

akas_df = pd.read_csv("akas.tsv", sep="\t")

akas_basics_df = basics_df.join(akas_df, $"tconst" === $"titleId")
akas_basics_df
```
